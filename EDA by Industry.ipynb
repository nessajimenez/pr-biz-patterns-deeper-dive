```python
df.naics=df.naics.astype('category')
```

```python
sns.lineplot(data=df,x='year',y="establishments",hue="naics",estimator='sum', errorbar=None)
```

```python
"""Because of the huge difference in numbers between metro area and the rest of the regions, it makes more sense to plot
and study them separately."""
sns.relplot(data=df,x='year',y="establishments", kind="line", hue="region", row="naics",errorbar=None, 
            height=5, aspect=1)
plt.show()
```

```python
mountain = df[df['region'] == 'Mountain']

eastern = df[df['region'] == 'Eastern']

metro = df[df['region'] == 'Metro Area']

southern = df[df['region'] == 'Southern']

northern = df[df['region'] == 'Northern']

western = df[df['region'] == 'Western']
```

```python
"""In the mountain region, Agriculture, Forestry, Fishing and Hunting seems to have had a very large increase from 
2016-2020 but in 2021 there was a large drop. 

Manufacturing is the only other industry to sharply increase in 2020, but it maintained those gains.

Retail Trade like department stores, warehouse clubs, gasoline, and niche stores saw a sharp decline in 2020 as well and
has not really recovered at all.

Accommodations, restaurants and beverage servicces remained steady.62


"""

plt.figure(figsize=[10,8])
sns.lineplot(data=mountain, x='year',y="establishments",style='naics', hue='naics',estimator='sum', errorbar=None, 
             palette='colorblind', markers=True)
plt.legend(title='NAICS', loc='upper left')
plt.title(label="Establishments in the Mountain Region by NAICS codes")
```

```python
plt.figure(figsize=[10,8])
sns.lineplot(data=southern, x='year',y="establishments",style='naics', hue='naics',estimator='sum', errorbar=None, 
             palette='colorblind', markers=True)
plt.legend(title='NAICS', loc='upper left')
plt.title(label="Establishments in the Mountain Region by NAICS codes")
```
