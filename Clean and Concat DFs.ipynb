```python
import pandas as pd
from Biz_Set_Functions import *
```

```python
#loading dfs, adding year column and standardizing column titles
df13 = made_df('Data Sets/cbp13pr_mun.csv', 2013)
df14 = made_df('Data Sets/cbp14pr_mun.csv', 2014)
df15 = made_df('Data Sets/cbp15pr_mun.csv', 2015)
df16 = made_df('Data Sets/cbp16pr_mun.csv', 2016)
df17 = made_df('Data Sets/cbp17pr_co.csv', 2017)
df18 = made_df('Data Sets/cbp_pr_mun_2018.csv', 2018)
df19 = made_df('Data Sets/cbp_pr_mun_2019.csv', 2019)

files = ['Data Sets/2020-1.csv','Data Sets/2020-2.csv','Data Sets/2020-3.csv','Data Sets/2020-4.csv']
df20 = combined_df(files, 2020)

files21 = ['Data Sets/_lmi_excel_QCEW_2021_Composición Industrial Municipio 1.csv',
           'Data Sets/_lmi_excel_QCEW_2021_Composición Industrial Municipio 2.csv',
          'Data Sets/_lmi_excel_QCEW_2021_Composición Industrial Municipio 3.csv',
          'Data Sets/_lmi_excel_QCEW_2021_Composición Industrial Municipio 4.csv']
df21 = combined_df(files21, 2021)

files22= ['Data Sets/2022_1.csv',
         'Data Sets/2022_2.csv',
         'Data Sets/2022_3.csv',
         'Data Sets/2022_4.csv']
df22 = combined_df(files22, 2022)
```

```python
#concat
df13_19 = pd.concat([df13,df14,df15,df16,df17,df18,df19])
df13_19

df20_22 = pd.concat([df20,df21,df22])
df20_22

#translating and renaming columns

def rename_columns(x):
    new_names={'industria':'industry','cty':'town_code','municipio':'town_name','unidades':'establishments',
               'empleo promedio':'average_employment','promedio':'average','est':'establishments','fipscty':'town_code'}
    x.columns=x.columns.str.strip()
    x.rename(columns=new_names, inplace=True)
    return x

df12_19 = rename_columns(df13_19)
df20_22 = rename_columns(df20_22)

#final concat
df13_22 = pd.concat([df13_19, df20_22], axis = 0)
df13_22
```

```python
#dropping unneeded columns


columns_to_drop = ['fipstate','empflag', 'emp_nf', 'emp', 'qp1_nf',
                   'qp1', 'ap_nf', 'ap', 'n1_4', 'n5_9', 'n10_19', 'n20_49',
                   'n50_99', 'n100_249', 'n250_499', 'n500_999', 'n1000', 'n1000_1',
                   'n1000_2', 'n1000_3', 'n1000_4', 'censtate', 'cencty', 'unnamed: 10', 'unnamed: 11',
                   'unnamed: 12', 'unnamed: 13', 'unnamed: 14', 'unnamed: 15',
                   'unnamed: 16', 'unnamed: 17', 'unnamed: 18', 'unnamed: 19',
                   'unnamed: 20', 'unnamed: 21', 'unnamed: 22', 'unnamed: 23',
                   'unnamed: 24', 'average_employment','salrio total municipio cuarto trimestre 2021',
                   'salrio total municipio tercer  trimestre 2021','salrio total municipio primer trimestre 2021',
                   'salrio total municipio primr trimestre 2022','salario total municipio segundo trimestre 2022',
                   'salario total municipio tercer trimestre 2022','average_employment',
                   'salrio total municipio segundo trimestre 2019','salario total','unnamed: 8','unnamed: 9',
                   'unnamed: 25','unnamed: 26','n<5','average','industry']

df13_22.drop(columns = columns_to_drop, inplace=True)

df13_22
```

```python
#updating column dtypes
df13_22["establishments"]=pd.to_numeric(df13_22["establishments"], errors='coerce')
df13_22['year'] = pd.to_datetime(df13_22['year'], format='%Y')
```

```python
#dropping null rows based on columns
df13_22.dropna(subset=['naics','establishments'], inplace=True)
df13_22.isna().sum()
```

```python
#dropping null rows based on having no useful data
no_towns_naics = df13_22[(df13_22['town_name'] == 'No Codificado') | (df13_22['naics'] == '------')].index
multi_town = df13_22[(df13_22['town_name'] == 'Multi Municipio') | (df13_22['town_name'] == 'Statewide')].index

df13_22.drop(no_towns_naics, inplace=True)
df13_22.drop(multi_town, inplace=True)

df13_22['town_name'].replace('nan',None,inplace=True)
```

```python
df13_22['town_name'].value_counts(dropna=False)
```

```python
#standardizing naics column. We just want to look at overall industry, not get into the nitty gritty subcategories

def replace_naics(naics_code):
    # Dictionary mapping NAICS codes to industry names
    naics_dict = {
    "11": "11","21": "21","22": "22","23": "23","31": "31","32": "32","33": "33","42": "42","44": "44","45": "45",
    "48": "48","49": "49","51": "51","52": "52","53": "53","54": "54","55": "55","56": "56","61": "61","62": "62","71":"71",
    "72": "72","81": "81","92": "92"
}
    
    industry_code = naics_code[:2]  # Extract first two digits of NAICS code
    return naics_dict.get(industry_code, "00")


df13_22['naics'] = df13_22['naics'].apply(replace_naics)
```

```python
#dropping rows of the bottom 10 industries that are tiny
df13_22 = df13_22[df13_22['naics'] != 61 | 51 | 71 | 49 | 21 | 55 | 22 | 11 | 92]

df13_22['naics'] = pd.to_numeric(df13_22['naics'])
```

```python
#we decided to group the towns by region to make graphing much more streamlined
df13_22['region'] = replace_town_with_region(df13_22,'town_code')

#making sure all towns labeled accurately
df13_22['town_name'] = replace_code_with_town(df13_22, 'town_code')

#resetting the index
df13_22 = df13_22.reset_index(drop=True)
```
